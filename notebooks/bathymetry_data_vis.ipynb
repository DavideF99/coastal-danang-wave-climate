```python
import xarray as xr
import matplotlib.pyplot as plt
from pathlib import Path

FIGURES_DIR = Path("../outputs/figures")
FIGURES_DIR.mkdir(parents=True, exist_ok=True)

bathy = xr.open_dataset("../data/GEBCO_data/gebco_2025_n18.0_s14.0_w106.0_e110.0.nc")
elev = bathy["elevation"]  # negative = below sea level

# Quick plot — mask land
import numpy as np
sea = elev.where(elev < 0)

fig, ax = plt.subplots(figsize=(7, 6))
sea.plot(ax=ax, cmap="Blues_r", vmin=-3000, vmax=0,
         cbar_kwargs={"label": "Depth (m)"})
ax.set_title("GEBCO 2025 — Da Nang Bay region")
plt.tight_layout()
plt.savefig(FIGURES_DIR / "bathymetry_data.png")
plt.show()
print("Saved → outputs/figures/bathymetry_data.png.png")
```
